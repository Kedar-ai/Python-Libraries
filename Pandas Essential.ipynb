# Reshaping

## stack() and unstack()

```python
import pandas as pd

df = pd.read_csv('olympics.csv',skiprows=4)
df.head()
```

- Athletes winning medals in Beijing Olympics 100m or 200m track event

```python
mw = df[(df.Edition == 2008) & ( (df.Event == '100m') | (df.Event == '200m'))]
mw
```

```python
g = mw.groupby(['NOC','Gender','Discipline','Event']).size()
g
```

```python
df = g.unstack(['Discipline','Event'])
df
```

- Plot the number of gold medals won by the USA male and female athletes throughout the history of the Olympics?

```python
gu = df[(df.NOC == 'USA') & (df.Medal == 'Gold')]
gu.groupby(['Edition','Gender']).size().unstack('Gender',fill_value=0).plot()
```

- Plot the 5 athletes who have won the most gold medals over the history of the Olympics. When there is a tie, consider the number of silver medals, then bronze medals.

```python
g = df.groupby(['Athlete','Medal']).size().unstack('Medal',fill_value=0)
g.sort_values(['Gold','Silver','Bronze'],ascending=False)[['Gold','Silver','Bronze']].head().plot(kind='bar')
```

# Challenge

## In every Olympics, which US athlete has won the most total number of medals? Include the athlete's discipline.

```python
gy  = df[df.NOC == 'USA']
gy = gy.groupby(['Edition','Athlete','Medal']).size().unstack('Medal',fill_value=0)
gy['Total'] = gy['Gold'] + gy['Silver'] + gy['Bronze']
gy.reset_index(inplace=True)
tu = [group.sort_values('Total',ascending=False)[:1] for year,group in gy.groupby('Edition')]
tu
top = pd.DataFrame()
for i in tu:
    top = top.append(i)
top
```
